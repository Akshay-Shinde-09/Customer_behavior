There are 37 null values in Review Rating. We can remove it using median value (But its not a best practice). 

Find median review rating within each category and replace it with missing values.

```python
df['Review Rating'] = df['Review Rating'].transform(lambda x: x.fillna(x.median))
```

```python
df.isnull().sum()
```

```python
# Sanke casing the coulumn names. To make it more neat, clear and consistent.
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# Feature enginerring. 
# Create age group. Using pd.qcut - it devides the data in to desirable parts (in this case 4) and then groups it accordingly.
label = ['Young Adult','Adult','Middle age','Senior']
df['age_group'] = pd.qcut(df['age'],q = 4, labels= label)
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].value_counts()
```

```python
# Create purchase_frequency_days column

frequncy_mapping = {
    'Every 3 Months':90,
    'Annually': 365,
    'Quarterly':90,
    'Monthly':30,
    'Bi-Weekly':14,
    'Fortnightly':14,
    'Weekly':7
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequncy_mapping)
df['purchase_frequency_days'].head(10)
```

```python
# Check if 'discount_applied' = 'promo_code_used'
(df['discount_applied'] == df['promo_code_used']).all()
```

Both columns hace same value so we drop 1 columns 'promo_code_used'. 

```python
df = df.drop('promo_code_used', axis= 1)
df = df.drop('review_rating', axis= 1) 
```
